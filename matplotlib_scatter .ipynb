```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1026 # data size
x = np.random.normal(0, 1, n) # 每個點的x值
y = np.random.normal(0, 1, n) # 每個點的y值
t = np.arctan2(y, x) # 顏色
```

```python
#s(點大小)、c(點顏色)、colormap(顏色樣式)、alpha(點透明度)
plt.scatter(x, y, s=75, c=t, cmap='plasma',alpha=0.5)

plt.xlim(-2, 2) # x, y軸範圍
plt.ylim(-2, 2)

plt.xticks(()) # x, y座標
plt.yticks(())

plt.show()
```
