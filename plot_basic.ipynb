```python
# import model
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 準備資料
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2
```

```python
# 畫圖

# 框出作圖範圍
plt.figure(num='plot', figsize=(8, 5))

# data fitting and line style
l1, = plt.plot(x, y2, color='blue', linewidth=1.0, label = 'liner line')
l2, = plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label = 'square line')

# 圖例和其位置
plt.legend(handles=[l1, l2], labels=['up', 'down'], loc='lower right')

# axis-範圍和名稱
plt.xlim(-1, 2)
plt.ylim(-2, 3)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# axis-刻度
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])

# axis-移動位置
# gca = get current axis
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

```

```python
# show()
plt.show()
```
